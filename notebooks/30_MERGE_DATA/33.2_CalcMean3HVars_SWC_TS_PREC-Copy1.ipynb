# Load data

```python
df = load_parquet(filepath="33.1_CH-CHA_IRGA+QCL+LGR+M10+MGMT_Level-1_eddypro_fluxnet_2005-2024.parquet")
df
```

# Testing mean settings to get mean over *preceding* records

```python
# Testing: closed='left' is the correct setting for preceding records
_df = df[['SWC_GF1_0.15_1_gfXG']].copy()
_df['mean_left'] = _df['SWC_GF1_0.15_1_gfXG'].rolling(window=6, closed='left').mean()
_df['mean_aright'] = _df['SWC_GF1_0.15_1_gfXG'].rolling(window=6, closed='right').mean()
display(_df['SWC_GF1_0.15_1_gfXG'].iloc[0:6].mean())
display(_df['SWC_GF1_0.15_1_gfXG'].iloc[1:7].mean())
display(_df.iloc[0:20])
```

# Calc 3-hour means from preceding records

```python
[print(c) for c in df.columns if "SWC_" in c];
[print(c) for c in df.columns if "TS_" in c];
[print(c) for c in df.columns if "PREC_" in c];
```

```python
var = ['SWC_GF1_0.15_1_gfXG', 'TS_GF1_0.04_1_gfXG', 'TS_GF1_0.15_1_gfXG', 'TS_GF1_0.4_1_gfXG', 'PREC_RAIN_TOT_GF1_0.5_1']

for v in var:    
    name_mean = f"{v}_MEAN3H"
    df[name_mean] = df[v].rolling(window=6, closed='left').mean()
```

```python
for v in var:
    print(f"\nResults for {v}: ")
    display(df[[v, f"{v}_MEAN3H"]].iloc[0:9])
```

# Save to file

```python
OUTNAME = "33.3_CH-CHA_IRGA+QCL+LGR+M10+MGMT_Level-1_eddypro_fluxnet_2005-2024"
filepath = save_parquet(filename=OUTNAME, data=df)
df.to_csv(f"{OUTNAME}.csv")
```
