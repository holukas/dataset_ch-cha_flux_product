# Imports

```python
import importlib.metadata
import pandas as pd
import warnings
from datetime import datetime
from pathlib import Path

from diive.core.io.filereader import search_files, MultiDataFileReader
from diive.core.io.files import save_parquet, load_parquet
# pd.options.display.width = None
# pd.options.display.max_columns = None
pd.set_option('display.max_rows', 3000)
pd.set_option('display.max_columns', 3000)
warnings.filterwarnings(action='ignore', category=FutureWarning)
warnings.filterwarnings(action='ignore', category=UserWarning)
dt_string = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
version_diive = importlib.metadata.version("diive")
print(f"diive version: v{version_diive}")
```

# **IRGA**: Load merged L1 results from parquet (available vars only)

```python
irga = load_parquet(filepath="../20_FLUXES_L1_IRGA_preparation/23.1_CH-CHA_IRGA_Level-1_eddypro_fluxnet_2005-2023_availableVars_meteo7.parquet")
display(irga.head(3))
display(irga.tail(3))
```

# **QCL+LGR**: Load merged L1 results from parquet (available vars only)

```python
qcllgr = load_parquet(filepath="../30_FLUXES_L1_IRGA_FluxProcessingChain/33.1_CH-CHA_IRGA_Level-1_eddypro_fluxnet_2005-2023_availableVars_noMeteo.parquet")
display(qcllgr.head(3))
display(qcllgr.tail(3))
```

# Identify variables

```python
print("Columns from QCL/LGR output that are not in the IRGA output:")
keepcols = [c for c in qcllgr.columns if c not in irga.columns]
print(f"{len(keepcols)} variables:")
keepcols
```

# Remove columns already available in IRGA data from QCL/LGR data

```python
print("Reduced QCL/LGR data:")
qcllgr_reduced = qcllgr[keepcols].copy()
display(qcllgr_reduced.describe())
```

# Merge IRGA data with QCL/LGR data

```python
irga_qcllgr = pd.concat([irga, qcllgr_reduced], axis=1)
```

```python
irga_qcllgr.describe()
```

# Adjust time range, last year 2023

```python
range = (irga_qcllgr.index.year >= 2005) & (irga_qcllgr.index.year <= 2023)
irga_qcllgr = irga_qcllgr[range].copy()
display(irga_qcllgr.head(3))
display(irga_qcllgr.tail(3))
```

# Save to file

```python
OUTNAME = "41.1_CH-CHA_IRGA_LGR+QCL_Level-1_eddypro_fluxnet_2005-2023_meteo7"
filepath = save_parquet(filename=OUTNAME, data=irga_qcllgr)
irga_qcllgr.to_csv(f"{OUTNAME}.csv")
```
