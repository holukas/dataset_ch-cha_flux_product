# Imports

```python
from pathlib import Path
import pandas as pd
from datetime import datetime
from diive.core.io.files import load_parquet, save_parquet
```

# Load results

```python
base = load_parquet(filepath=Path("../30_MERGE_DATA") / "33.1_CH-CHA_IRGA+QCL+LGR+M7+MGMT_Level-1_eddypro_fluxnet_2005-2023.parquet")
additional_meteo = load_parquet(filepath=Path("../10_METEO") / "17.1_CH-CHA_addMeteo_PREC_SWC_TS_2005-2023.parquet")
nee = load_parquet(filepath=Path("../40_FLUX_PROCESSING_CHAIN") / "41.1_FluxProcessingChain_L4.1_NEE.parquet")
le = load_parquet(filepath=Path("../40_FLUX_PROCESSING_CHAIN") / "42.1_FluxProcessingChain_L4.1_LE.parquet")
h = load_parquet(filepath=Path("../40_FLUX_PROCESSING_CHAIN") / "43.1_FluxProcessingChain_L4.1_H.parquet")
```

```python
display(base)
display(additional_meteo)
display(nee)
display(le)
display(h)
```

# Merge files

## Start with base data

```python
df = base.copy()
df
```

## Add *new* columns from flux processing chain

```python
newcols = [c for c in nee.columns if c not in df.columns]
[print(f"++ Adding variable {c} from flux processing chain") for c in newcols]
df = pd.concat([df, nee[newcols]], axis=1)
df
```

```python
newcols = [c for c in le.columns if c not in df.columns]
[print(f"++ Adding variable {c} from flux processing chain") for c in newcols]
df = pd.concat([df, le[newcols]], axis=1)
df
```

```python
newcols = [c for c in h.columns if c not in df.columns]
[print(f"++ Adding variable {c} from flux processing chain") for c in newcols]
df = pd.concat([df, h[newcols]], axis=1)
df
```

## Add *new* columns from additional meteo data

```python
newcols = [c for c in additional_meteo.columns if c not in df.columns]
[print(f"++ Adding variable {c} from additional meteo data") for c in newcols]
df = pd.concat([df, additional_meteo[newcols]], axis=1)
df
```

```python
df.index
```

# Save to file

```python
OUTNAME = "51.1_CH-CHA_IRGA-L4.1+QCL-L1+LGR-L1+M10+MGMT_2005-2023"
filepath = save_parquet(filename=OUTNAME, data=df)
df.to_csv(f"{OUTNAME}.csv")
```
